```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
transactions = pd.read_csv("clean_transactions.csv")
```

```python
transactions['Discount'] = transactions['Price'] - transactions['Selling_price']
```

```python
category_sales = transactions.groupby('P_CATEGORY')['Selling_price'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
category_sales.plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
Insight

The analysis shows that the DECOR category generated the highest
total sales among all product categories.
This indicates that consumers/customers are highly
interested in decorative products compared to other categories.
```

```python
payment_discount = transactions.groupby('Payment Method')['Discount'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
payment_discount.plot(kind='bar')
plt.title('Total Discount by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Discount')
plt.show()
```

```python
Insight

The Payment Method Mobile Carrier Billing received the
Highest total discount compared to other payment methods.
This suggests that merchants may provide higher promotional
offers for this payment option to encourage its usage.
```

```python
condition_counts = transactions['CONDTION'].value_counts()
```

```python
plt.figure(figsize=(6,6))
condition_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Product Condition Distribution')
plt.ylabel('')
plt.tight_layout()
plt.savefig('product_condition_pie_chart.png')
plt.show()
```

```python
Insight

The majority of the customers preferred purchasing
new products,followed by used and refurbished products.
This indicates that customers/consumers generally
prefer brand-new items,but there is also a noticable demand
for lower-prized refurbished or used products.
```

```python
returns = transactions['Return_ind'].value_counts()
```

```python
plt.figure(figsize=(6,4))
returns.plot(kind='bar')
plt.title('Product Returns')
plt.xlabel('Return Indicator')
plt.ylabel('Count')
plt.show()
```

```python
Insight

The analysis shows that most products were not returned,
while a smaller portion of products were returned.
This suggests that the majority of customers are
satisfied with their purchases.
```

```python
Conclusion:

In this project, we have analyzed customer transactions and product sales data
to identify key purchasing patterns.The analysis revealed that the Decor
category generated the highest sales,and Mobile Carrier Billing offered
the highest disounts. Most of the customers/consumers preffered purchasing
new products,and the majority of the transactions resulted in successful
purchases without returns. These insights help understand customer
behaviour and sales trends, which can assist businesses in improving their
marketing strategies and product offerings.
```
